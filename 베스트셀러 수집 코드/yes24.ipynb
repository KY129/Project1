```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.yes24.com/Product/Category/MonthWeekBestSeller?categoryNumber=001&pageNumber=1&pageSize=40&type=month&saleYear=2024&saleMonth=11"
res = requests.get(url)

books_data = []

if res.status_code == 200:
    soup = BeautifulSoup(res.text, "html.parser")
    trs = soup.select("#yesBestList .itemUnit")

    rank = 1
    for item in trs:
        if rank > 30:
            break
        
        title = item.select_one(".gd_name").get_text(strip=True)
        href = item.find("a")["href"]
        auth_span = item.select_one('span.authPub.info_auth')
        a_tag = auth_span.find('a')
        author = a_tag.text.strip()
        
        url2 = "https://www.yes24.com" + href
        res2 = requests.get(url2)
        
        if res2.status_code == 200:
            book_soup = BeautifulSoup(res2.text, "html.parser")
            links = book_soup.select(".infoSetCont_wrap > dl > dd > ul > li > a")
            if links:
                genre = links[1].get_text(strip=True)
                genre = genre.replace("에세이", "소설/시/희곡")
                genre = genre.replace("소설/시/희곡", "소설/시/희곡/에세이")
                genre = genre.replace("경제 경영", "경제/경영")
              
        
        books_data.append({
            "순위": rank,
            "제목": title,
            "작가": author,
            "장르": genre
        })
        
        rank += 1

books_df = pd.DataFrame(books_data)
books_df
```
